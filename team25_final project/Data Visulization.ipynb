```python
for col in set(breast_cancer.columns) - set(breast_cancer.describe().columns):
    breast_cancer[col] = breast_cancer[col].astype('category')
for i,j in zip(breast_cancer.columns,(breast_cancer.values.astype(str) == '?').sum(axis = 0)):
    if j > 0:
        print(str(i) + ': ' + str(j) + ' records')
```

```python
newData = breast_cancer.replace({'?':pd.NaT}).dropna()
newData.shape
```

```python
newData.tail(40)
```

```python
display(newData.corr())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

size = newData['class'].value_counts()
print(size)
colors = ['lightgreen', 'orange']
labels = "no-recurrence-events", "recurrence-events"
explode = [0, 0.01]

data_circle = plt.Circle((0, 0), 0.7, color = 'white') 

plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size, colors = colors, labels = labels, shadow = True, explode = explode, autopct = '%.2f%%')
plt.title('Distribution of recurrence', fontsize = 18)
p = plt.gcf()
p.gca().add_artist(data_circle)
plt.legend()
plt.show()
```

```python
dat = pd.crosstab(newData['age'], newData['class'])
age = dat.div(dat.sum(1).astype(float), axis = 0)
fig = age.plot(kind ='bar', color = ['green', 'orange'])
plt.title('Relation between age and recurrence', fontsize = 18)
fig.set_xlabel('age')
fig.set_ylabel('Proportion of recurrence')
plt.show()
```
